```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

sns.set(style="whitegrid")

years = [2024, 2025]
world_growth = [3.2, 2.9]
us_growth = [2.1, 1.6]
china_growth = [4.8, 4.2]
indonesia_growth_range = (4.7, 5.5)
indonesia_growth_est = 4.9

exports_data = {
    "Negara": ["AS", "Tiongkok", "ASEAN", "Jepang"],
    "Ekspor (Miliar USD)": [18.5, 27.3, 36.2, 14.1]
}
exports_df = pd.DataFrame(exports_data)

regions = ["Jawa", "Kalimantan", "Sumatera", "Sulawesi", "Papua"]
growth_rate = [5.1, 5.3, 4.8, 4.6, 3.9]

fig, axs = plt.subplots(2, 2, figsize=(14, 10))

axs[0, 0].plot(years, world_growth, marker='o', label="Dunia")
axs[0, 0].plot(years, us_growth, marker='o', label="AS")
axs[0, 0].plot(years, china_growth, marker='o', label="Tiongkok")
axs[0, 0].set_title("Pertumbuhan Ekonomi Global (%)")
axs[0, 0].set_xlabel("Tahun")
axs[0, 0].set_ylabel("Persentase")
axs[0, 0].legend()

sns.barplot(x="Negara", y="Ekspor (Miliar USD)", data=exports_df, ax=axs[0, 1], palette="Blues_d")
axs[0, 1].set_title("Ekspor Nonmigas Indonesia Triwulan I 2025")

sns.barplot(x=growth_rate, y=regions, ax=axs[1, 0], palette="Greens_d")
axs[1, 0].set_title("Pertumbuhan Ekonomi per Wilayah Indonesia (%)")

axs[1, 1].barh(["Proyeksi BI 2025"], [indonesia_growth_est], color='orange')
axs[1, 1].axvspan(indonesia_growth_range[0], indonesia_growth_range[1], color='gray', alpha=0.3, label='Kisaran BI (4.7% - 5.5%)')
axs[1, 1].set_xlim(4.5, 5.7)
axs[1, 1].set_title("Proyeksi Pertumbuhan Ekonomi Indonesia 2025")
axs[1, 1].legend()

plt.tight_layout()
plt.show()
```
